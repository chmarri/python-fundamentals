```python
def average(*values):
    try:
        return sum(values) / len(values)
    except ZeroDivisionError:
        return 0
    
average(1,2,3)
```

```python
def average (*values):
    if len(values) == 0:
        return 0
    return sum(values) / len(values)

average(1,345,2378,9872348907234)
```

```python
def product(*values):
    prod =1 
    for value in values:
        prod *= value
    return prod
product(1,2,3,4)
```
