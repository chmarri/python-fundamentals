```python
def identity(rows, cols):
    return[
        [1 if row == col else 0 for col in range(cols)]
        for row in range(rows)
    ]
    
identity(5,5)
```

```python
f = lambda rows, cols: [
        [1 if row == col else 0 for col in range(cols)]
        for row in range(rows)
    ]
f(5,5)
```
