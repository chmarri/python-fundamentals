```python
def find_max(a,b,c):
    max_ = a
    if b > max_:
        max_ = b
    if c > max_:
        max_ = c
    return max_

find_max(10,20,30)
       
            
```

```python
from datetime import datetime, timezone

def log(message):
    curr_time = datetime.now(timezone.utc).isoformat()
    print(f'{curr_time} - [{message}]')
    
# log('log 1')
result = log('log 1')
```

```python
def is_all_positive(data):
    for element in data:
        if element < 0:
            return False
    return True

is_all_positive([1,2,3,4,5])
```

```python
is_all_positive([1,2,3,4,5, -4])
```

```python
def gen_matrix(m, n, default_value):
    return [[default_value for i in range(n)] for j in range(m)]

gen_matrix(8,16,1)
```

```python
def gen_matrix(rows, cols, default_value):
    return [[default_value for i in range(cols)] for j in range(rows)]

gen_matrix(8,16,1)
```
