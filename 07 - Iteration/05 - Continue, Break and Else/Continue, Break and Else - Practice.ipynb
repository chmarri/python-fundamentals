```python
for i in range (1,5):
    for j in range (1,5):
        if (i + j) % 2 == 1:
            print (f'{i} + {j} is odd, skipping...')
            continue
        print ( f'adding number: {i} + {j} = {i + j}')
```

```python
# Processing orders - skip orders that need special handling
orders = ['Standard', 'Express', 'Standard', 'Overnight']
customers = ['Alice', 'Bob', 'Charlie', 'Diana']

for customer in customers:
    for order in orders:
        if order == 'Express' or order == 'Overnight':
            print(f'{customer} - {order}: needs special handling, skipping...')
            continue
        print(f'Processing {order} order for {customer}')
```

```python
# Processing files - skip certain file types
files = ['report.pdf', 'data.csv', 'image.jpg', 'notes.txt']
users = ['John', 'Sarah', 'Mike']

for user in users:
    for file in files:
        if file.endswith('.jpg') or file.endswith('.pdf'):
            print(f'{user} - {file}: binary file, skipping...')
            continue
        print(f'{user} is processing {file}')
```
