```python
## Example 2: TypeError - Invalid Type for Operation
```

```python
def calculate_discount(price, discount_percent):
    """Calculate discounted price. Discount must be a number."""
    if not isinstance(discount_percent, (int, float)):
        raise TypeError(f"Discount must be a number, not {type(discount_percent).__name__}")
    
    if discount_percent < 0 or discount_percent > 100:
        raise ValueError("Discount must be between 0 and 100")
    
    return price - (price * discount_percent / 100)

# Test with valid input
print(calculate_discount(100, 20))  # Works: 80.0

# Test with invalid type - will raise TypeError
# print(calculate_discount(100, "20%"))  # Uncomment to see error
```

```python
## Example 3: IndexError - Invalid List Index
```

```python
def get_student_grade(grades, student_number):
    """Get grade for a student by their number (1-based index)."""
    # Convert to 0-based index
    index = student_number - 1
    
    if index < 0:
        raise IndexError("Student number must be positive")
    
    if index >= len(grades):
        raise IndexError(f"Student number {student_number} not found. Only {len(grades)} students in list.")
    
    return grades[index]

grades = ['A', 'B', 'C', 'A', 'B']

# Valid access
print(get_student_grade(grades, 3))  # Works: C (3rd student)

# Invalid access - will raise IndexError
# print(get_student_grade(grades, 10))  # Uncomment to see error
```

```python
## Example 4: KeyError - Missing Dictionary Key
```

```python
def get_user_info(user_data, required_fields):
    """Validate that all required fields exist in user data."""
    missing_fields = []
    
    for field in required_fields:
        if field not in user_data:
            missing_fields.append(field)
    
    if missing_fields:
        raise KeyError(f"Missing required fields: {', '.join(missing_fields)}")
    
    return user_data

# Valid user data
user1 = {'name': 'Alice', 'email': 'alice@example.com', 'age': 25}
required = ['name', 'email', 'age']
print(get_user_info(user1, required))  # Works

# Invalid user data - missing fields
user2 = {'name': 'Bob'}
# print(get_user_info(user2, required))  # Uncomment to see error - missing email and age
```

```python
## Summary of Exception Types

- **ValueError**: Invalid value (right type, wrong value)
- **TypeError**: Wrong data type
- **IndexError**: Index out of range in a sequence
- **KeyError**: Key doesn't exist in dictionary
- **ZeroDivisionError**: Division by zero
- **FileNotFoundError**: File doesn't exist
- **AttributeError**: Attribute doesn't exist on object
```

```python
import os

def read_config_file(filename):
    """Read configuration file. Raise error if file doesn't exist."""
    if not os.path.exists(filename):
        raise FileNotFoundError(f"Configuration file '{filename}' not found. Please create it first.")
    
    with open(filename, 'r') as f:
        return f.read()

# This will raise FileNotFoundError
# print(read_config_file('config.txt'))  # Uncomment to see error

# To test with a file that exists, uncomment below:
# with open('test_config.txt', 'w') as f:
#     f.write('sample config')
# print(read_config_file('test_config.txt'))
```

```python
## Example 6: AttributeError - Missing Attribute
```

```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    
    def get_attribute(self, attr_name):
        """Get car attribute, raise error if it doesn't exist."""
        if not hasattr(self, attr_name):
            raise AttributeError(f"Car object has no attribute '{attr_name}'. Available: brand, model")
        return getattr(self, attr_name)

my_car = Car('Toyota', 'Camry')

# Valid attribute
print(my_car.get_attribute('brand'))  # Works: Toyota

# Invalid attribute - will raise AttributeError
# print(my_car.get_attribute('color'))  # Uncomment to see error
```

```python
## Example 7: AssertionError - Assertion Fails
```

```python
def withdraw_money(balance, amount):
    """Withdraw money from account. Balance must not go negative."""
    if amount <= 0:
        raise ValueError("Withdrawal amount must be positive")
    
    if amount > balance:
        raise AssertionError(f"Insufficient funds! Balance: ${balance}, Requested: ${amount}")
    
    return balance - amount

# Valid withdrawal
print(f"New balance: ${withdraw_money(100, 30)}")  # Works: $70

# Invalid withdrawal - insufficient funds
# print(withdraw_money(100, 150))  # Uncomment to see error
```

```python
## Example 8: RuntimeError - General Runtime Issue
```

```python
class Database:
    def __init__(self):
        self.connected = False
    
    def connect(self):
        self.connected = True
        print("Database connected")
    
    def query(self, sql):
        """Execute query. Must be connected first."""
        if not self.connected:
            raise RuntimeError("Database not connected. Call connect() first.")
        return f"Executing: {sql}"

db = Database()

# This will raise RuntimeError - not connected
# print(db.query("SELECT * FROM users"))  # Uncomment to see error

# Proper usage
db.connect()
print(db.query("SELECT * FROM users"))  # Works
```

```python
def calculate_with_variable(expression, variables):
    """Evaluate expression with given variables. Check if all variables are defined."""
    required_vars = ['x', 'y']  # Variables needed for expression
    
    for var in required_vars:
        if var not in variables:
            raise NameError(f"Variable '{var}' is not defined. Available: {list(variables.keys())}")
    
    # Evaluate expression
    x = variables['x']
    y = variables['y']
    return eval(expression)

# Valid - all variables provided
result = calculate_with_variable('x + y', {'x': 10, 'y': 20})
print(f"Result: {result}")  # Works: 30

# Invalid - missing variable y
# result = calculate_with_variable('x + y', {'x': 10})  # Uncomment to see error
```
