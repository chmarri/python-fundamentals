```python
names = ['alice', 'bob', 'charlie', 'diana', 'eve']
```

```python
# Traditional for loop approach
uppercase_names = []
for name in names:
    uppercase_name = name.upper()
    uppercase_names.append(uppercase_name)

uppercase_names
```

```python
# List comprehension approach - much shorter!
uppercase_names = [name.upper() for name in names]
uppercase_names
```

```python
# Using the names list from earlier
names = ['alice', 'bob', 'charlie', 'diana', 'eve', 'frank', 'grace', 'henry']
```

```python
# Traditional for loop - uppercase all names
start = perf_counter()
for i in range(1_000_000):
    uppercase_names = []
    for name in names:
        uppercase_name = name.upper()
        uppercase_names.append(uppercase_name)
end = perf_counter()

elapsed_time = end - start
print(uppercase_names)
print(f"For loop took: {elapsed_time:.4f} seconds")
```

```python
# List comprehension - uppercase all names
start = perf_counter()
for i in range(1_000_000):
    uppercase_names = [name.upper() for name in names]
end = perf_counter()

elapsed_time = end - start
print(uppercase_names)
print(f"List comprehension took: {elapsed_time:.4f} seconds")
```

```python
### Another Example: Creating Email Addresses
```

```python
# Traditional for loop - create email addresses
start = perf_counter()
for i in range(1_000_000):
    emails = []
    for name in names:
        email = f"{name}@example.com"
        emails.append(email)
end = perf_counter()

elapsed_time = end - start
print(emails)
print(f"For loop took: {elapsed_time:.4f} seconds")
```

```python
# List comprehension - create email addresses
start = perf_counter()
for i in range(1_000_000):
    emails = [f"{name}@example.com" for name in names]
end = perf_counter()

elapsed_time = end - start
print(emails)
print(f"List comprehension took: {elapsed_time:.4f} seconds")
```
