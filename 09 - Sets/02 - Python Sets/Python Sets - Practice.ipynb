```python
# Set Difference: s1 - s2 returns elements in s1 but NOT in s2
s1 = {1, 2, 3}
s2 = {3, 4, 5}
print("s1:", s1)
print("s2:", s2)
print("s1 - s2:", s1 - s2)  # Elements in s1 but not in s2
print("s2 - s1:", s2 - s1)  # Elements in s2 but not in s1
```

```python
# Alternative method using .difference()
print(s1.difference(s2))  # Same as s1 - s2
print(s2.difference(s1))  # Same as s2 - s1
```

```python
# Real world example: Finding unique items
my_skills = {'Python', 'JavaScript', 'SQL', 'Git'}
job_requirements = {'Python', 'Java', 'Docker', 'Git'}

skills_to_learn = job_requirements - my_skills
print("Skills I need to learn:", skills_to_learn)
```
