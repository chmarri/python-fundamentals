## üéØ Set Operations Summary

Understanding when to use each set operation:

**1. Union (`|` or `.union()`)** - Combine everything
- Use when: You want ALL unique items from multiple sets
- Example: Merging two lists of items

**2. Intersection (`&` or `.intersection()`)** - Find common elements
- Use when: You want items that appear in ALL sets
- Example: Finding mutual friends, common skills

**3. Difference (`-` or `.difference()`)** - What's unique to first set
- Use when: You want items in set A but NOT in set B
- Example: Items to learn, tasks remaining, items not returned

**4. Symmetric Difference (`^` or `.symmetric_difference()`)** - Either/or but not both
- Use when: You want items that are in one set OR the other, but NOT in both
- Example: Finding unique preferences between two people

````python
# Real-world example: Organization hierarchy
# Demonstrating subset/superset relationships

# Define permission levels
viewer = {'read'}
editor = {'read', 'write'}
moderator = {'read', 'write', 'delete'}
admin = {'read', 'write', 'delete', 'admin', 'manage_users'}

print("ORGANIZATION PERMISSION HIERARCHY")
print("=" * 50)
print(f"Viewer:    {viewer}")
print(f"Editor:    {editor}")
print(f"Moderator: {moderator}")
print(f"Admin:     {admin}")
print("=" * 50)

# Check hierarchy relationships
print("\nHierarchy Checks (Subset):")
print(f"Viewer ‚äÜ Editor?     {viewer.issubset(editor)}")  # True
print(f"Editor ‚äÜ Moderator?  {editor.issubset(moderator)}")  # True
print(f"Moderator ‚äÜ Admin?   {moderator.issubset(admin)}")  # True

print("\nReverse Hierarchy (Superset):")
print(f"Admin ‚äá Moderator?   {admin.issuperset(moderator)}")  # True
print(f"Admin ‚äá Editor?      {admin.issuperset(editor)}")  # True
print(f"Admin ‚äá Viewer?      {admin.issuperset(viewer)}")  # True

print("\nStrict Comparison:")
print(f"Viewer < Editor?     {viewer < editor}")  # True - proper subset
print(f"Editor < Editor?     {editor < editor}")  # False - equal sets
````

## üîç Subset & Superset Operations Explained

**Subset** means "contained within":
- `A.issubset(B)` or `A <= B` ‚Üí Is every element of A also in B?
- `A < B` (strict) ‚Üí Is A contained in B AND A ‚â† B?

**Superset** means "contains":
- `A.issuperset(B)` or `A >= B` ‚Üí Does A contain all elements of B?
- `A > B` (strict) ‚Üí Does A contain B AND A ‚â† B?

**Disjoint** means "no overlap":
- `A.isdisjoint(B)` ‚Üí Do A and B have NO common elements?

````python
# Demonstrating common pitfalls

# PITFALL 1: Sets are unordered
print("PITFALL 1: Order is NOT preserved")
s1 = {5, 2, 8, 1, 9}
s2 = {5, 2, 8, 1, 9}
print(f"Created: {{5, 2, 8, 1, 9}}")
print(f"Actually stored: {s1}")
print(f"Equal? {s1 == s2} (order doesn't matter)")
print()

# PITFALL 2: Mutable items not allowed
print("PITFALL 2: Only immutable items allowed")
try:
    bad_set = {[1, 2, 3]}  # Lists are mutable - ERROR!
except TypeError as e:
    print(f"‚úó Cannot add list: {e}")

good_set = {(1, 2, 3), 'hello', 42}  # Tuples, strings, numbers OK
print(f"‚úì Valid set with immutable items: {good_set}")
print()

# PITFALL 3: A - B ‚â† B - A
print("PITFALL 3: Subtraction is NOT commutative")
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
print(f"A = {a}")
print(f"B = {b}")
print(f"A - B = {a - b}  (unique to A)")
print(f"B - A = {b - a}  (unique to B)")
print(f"These are DIFFERENT! Order matters in subtraction.")
````

## üí° Common Pitfalls & Best Practices

### Pitfall 1: Sets don't preserve order
```python
# Order is not guaranteed
s = {3, 1, 2}  # May display as {1, 2, 3} or any order
```

### Pitfall 2: Sets only contain immutable items
```python
# ‚úì OK: {1, 2, 'hello', (1, 2)}
# ‚úó ERROR: {[1, 2], {'key': 'value'}}
```

### Pitfall 3: Difference is not commutative
```python
A - B ‚â† B - A  # Order matters!
```

### Best Practice: Use method names for clarity
```python
# Both work, but method names are more readable
s1.intersection(s2)  # Clearer
```

````python
# Step-by-step breakdown for clarity
import string

text = 'Chandra Mohan is a great person who worked he ass of for the family'

print("STEP-BY-STEP BREAKDOWN:")
print("=" * 60)

# Step 1
print("\n1. Get all lowercase letters from alphabet:")
all_letters = set(string.ascii_lowercase)
print(f"   {all_letters}")
print(f"   Total: {len(all_letters)} letters")

# Step 2
print("\n2. Convert text to lowercase:")
text_lower = text.lower()
print(f"   '{text_lower}'")

# Step 3
print("\n3. Remove spaces and convert to set:")
text_set = set(text_lower.replace(" ", ""))
print(f"   {text_set}")
print(f"   Note: This includes punctuation and letters")

# Step 4
print("\n4. Filter only alphabet letters from text:")
text_letters = text_set & all_letters  # Intersection removes non-letters
print(f"   {sorted(text_letters)}")

# Step 5
print("\n5. Find missing letters (alphabet - text_letters):")
missing = all_letters - text_letters
print(f"   {sorted(missing)}")
print(f"   Missing: {', '.join(sorted(missing))}")
print("=" * 60)
````

````python
# Practical function to check any sentence
def find_missing_letters(sentence):
    """Find which letters of the alphabet are missing from a sentence"""
    import string
    
    alphabet = set(string.ascii_lowercase)
    sentence_letters = set(sentence.lower()) & alphabet  # Only keep letters
    missing = alphabet - sentence_letters
    
    return {
        'missing': sorted(missing),
        'present': sorted(sentence_letters),
        'missing_count': len(missing),
        'coverage_percentage': (len(sentence_letters) / 26) * 100
    }

# Test with your sentence
text = 'Chandra Mohan is a great person who worked he ass of for the family'
result = find_missing_letters(text)

print(f"Sentence: {text}")
print(f"\nMissing letters ({result['missing_count']}): {', '.join(result['missing'])}")
print(f"Present letters ({len(result['present'])}): {', '.join(result['present'])}")
print(f"Coverage: {result['coverage_percentage']:.1f}% of alphabet")

# Test with a pangram (sentence with all letters)
pangram = "The quick brown fox jumps over the lazy dog"
result2 = find_missing_letters(pangram)
print(f"\n\nPangram test: {pangram}")
print(f"Missing: {result2['missing'] if result2['missing'] else 'None - Perfect pangram!'}")
````
